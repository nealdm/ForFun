```python
class fake_stock():
    '''
    # (number of points you would like, number of sinosoidal terms, and degree of linear terms involved) 
    # and it produces a vector of length n that approximates a 'noisy' vector that has 'some pattern' incorporated.
    # It would be cool to pull from cos(x)**(2.4 or something) between -pi/2 and pi/2 and put that number into tangent
    # and use that final output as reasonable coefficients for the stuff
    '''
    def __init__(self,seq_length,num_poly_terms,poly_width_and_offset):#,num_sin_terms,sin_width_and_offset):
        '''
        seq_length is the number of values in the data
        poly_coef_vec is the vector that contains the polynomial's coefficient values
        poly_width_and_offset is a tuple of vectors such as ([a0,b0,c0],[a1,b1,c1]) 
            where the first vector represents the width of the ith coefficient random number
            and the second vector represents the offset of the ith coefficient random number
        sin_phase_vec is the vector that contains the sin terms' phase values (phi)
        sin_freq_vec is the vector that contains the sin term's frequency values (w)
        sin_amp_vec is the vector that contains the sin term's amplitude values (A)
        (recall that the genric sin function is A*sin(w*t+phi))
        '''
        self.seq_length = seq_length
        self.poly_coef_vec = np.random.random((num_poly_terms))
        
        # scale and shift each term the desired ammount
        for i in range(num_poly_terms):
            self.poly_coef_vec[i] = self.poly_coef_vec[i]*poly_width_and_offset[0][i]+poly_width_and_offset[1][i]
#         self.sin_phase_vec = np.random.random((num_sin_terms))*2*np.pi
#         self.sin_freq_vec = np.random.random((num_sin_terms))
#         self.sin_amp_vec = np.random.random((num_sin_terms))
        
        # poly containes the polynomial terms of the fake stock
        self.poly = np.polynomial.polynomial.Polynomial(self.poly_coef_vec)
        # noise_vals contains the noise of the stock
        self.noise_vals = np.random.random((self.seq_length))
        
        function = np.vectorize(self.poly)
        
        fun_vals = function(np.linspace(0,self.seq_length,self.seq_length))
        self.vector = fun_vals+self.noise_vals
        
```

```python
def fit_sin(tt, yy):
    '''Fit sin to the input time sequence, and return fitting parameters "amp", "omega", "phase", "offset", "freq", "period" and "fitfunc"'''
    tt = numpy.array(tt)
    yy = numpy.array(yy)
    ff = numpy.fft.fftfreq(len(tt), (tt[1]-tt[0]))   # assume uniform spacing
    Fyy = abs(numpy.fft.fft(yy))
    guess_freq = abs(ff[numpy.argmax(Fyy[1:])+1])   # excluding the zero frequency "peak", which is related to offset
    guess_amp = numpy.std(yy) * 2.**0.5
    guess_offset = numpy.mean(yy)
    guess = numpy.array([guess_amp, 2.*numpy.pi*guess_freq, 0., guess_offset])

    def sinfunc(t, A, w, p, c):  return A * numpy.sin(w*t + p) + c
    popt, pcov = scipy.optimize.curve_fit(sinfunc, tt, yy, p0=guess)
    A, w, p, c = popt
    f = w/(2.*numpy.pi)
    fitfunc = lambda t: A * numpy.sin(w*t + p) + c
    return {"amp": A, "omega": w, "phase": p, "offset": c, "freq": f, "period": 1./f, "fitfunc": fitfunc, "maxcov": numpy.max(pcov), "rawres": (guess,popt,pcov)}
```

```python
def least_squares(A, b):
    """Calculate the least squares solutions to Ax = b by using the QR
    decomposition.

    Parameters:
        A ((m,n) ndarray): A matrix of rank n <= m.
        b ((m, ) ndarray): A vector of length m.

    Returns:
        x ((n, ) ndarray): The solution to the normal equations.
    """
    Q,R = la.qr(A,mode='economic') # This is all straight out of the book.
    return la.solve_triangular(R, Q.T @ b)

def line_fit(data):
    """Find the least squares line that fits to the datapoints given. Plot both the data points and the least
    squares line.
    """
    A = np.column_stack((data[:,0],np.ones(len(data))))
    b = data[:,1]
    # find least sq solution
    x = least_squares(A,b)
    x_line = np.linspace(min(data[:,0]),max(data[:,0]),250)
    # plot data points as scatterplot and least sq line
    plt.title("Datapoints with least squares line")
    plt.plot(data[:,0],data[:,1],'k.')
    plt.plot(x_line,x[0]*x_line+x[1],'b-')
    plt.xlabel("time")
    plt.ylabel("Price Index")
    plt.show()

    return

# Yup... that is all from ACME homework :D
```

```python
def find_line_from_period(sequence,period,offset):
    num_points = int(len(sequence)/period)
    positions = [period*i + offset for i in range(num_points)]
    values = sequence[int(offset)::int(period)]
    pairs = [[positions[i],values[i]] for i in range(len(positions))]
    print("pairs{}".format(pairs))
    line_fit(np.array(pairs))
    
# I can't remember what, but I think there is a problem with the period here- It should be how many ticks in
#  between the "usable" points.
```
